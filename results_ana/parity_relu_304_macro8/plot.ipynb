```python
import numpy as np
import matplotlib.pyplot as plt

# Original BSA data
years = np.array([1986, 1987, 1991, 1994, 1996, 1997, 2000, 2002, 2005, 2009, 2013, 2017, 2021, 2023])
trust_pct = np.array([40, 37, 33, 24, 22, 25, 16, 26, 26, 16, 17, 20, 24, 14])

# --- To change the aspect ratio, modify the figsize tuple (width, height) ---
# For example, a 16:9 ratio could be figsize=(8, 4.5)
# A square ratio would be figsize=(6, 6)
plt.figure(figsize=(7, 3.4)) # Changed to an 8:5 ratio

# --- The "wiggle" code has been removed ---
# We now plot the original data directly for a straight line between points.
plt.plot(years, trust_pct, color="black", linewidth=2, zorder=2) # zorder places line behind dots
plt.scatter(years, trust_pct, color="black", zorder=3)

# X-axis ticks only first and last year
plt.xticks([years[0], years[-1]], [str(years[0]), str(years[-1])])

# Y-axis ticks only min and max of original data
y_min, y_max = trust_pct.min(), trust_pct.max()
plt.yticks([y_min, y_max], [f"{y_min}%", f"{y_max}%"])

# Labels
plt.xlabel("Year")
plt.ylabel("Trust")

# Minimal y-grid
plt.grid(axis='y', linestyle="--", alpha=0.4)

plt.tight_layout()

# --- To save with a transparent background, use transparent=True ---
# Note: You must save the figure to see the transparency. It won't show in the display window.
# The filename should typically be .png to support transparency.
plt.savefig('transparent_plot2.png', transparent=True)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Pew Research Center – % who say officials don't care what people like them think
countries = ["Spain", "U.S.", "France", "U.K.", "Germany", "24-country\nmedian"]
values    = [85, 83, 74, 70, 63, 74]

fig, ax = plt.subplots(figsize=(8, 5), facecolor="none")   # transparent fig

# Bars in solid black
bars = ax.bar(countries, values, color="black", edgecolor="black")

# --- Axes formatting ------------------------------------
ax.set_ylabel("Percentage (%)")        # y-axis label only
ax.set_ylim(0, 100)
ax.set_yticks([0, 20, 40, 60, 80, 100])

# Keep x-tick labels (country names)
ax.set_xticks(range(len(countries)))
ax.set_xticklabels(countries)

# Remove all spines except bottom baseline
for key, spine in ax.spines.items():
    spine.set_visible(key == "bottom")

# --- Value labels ---------------------------------------
for bar, val in zip(bars, values):
    ax.text(bar.get_x() + bar.get_width()/2, val + 2,
            f"{val}%", ha="center", va="bottom")

# Transparent plot area
ax.patch.set_alpha(0)

# Optional title (delete if you prefer absolute minimalism)
ax.set_title("Feeling Ignored by Elected Officials – Selected Countries (Pew, 2024)")
plt.savefig('transparent_plot5.png', transparent=True)
plt.tight_layout()
plt.show()

```
