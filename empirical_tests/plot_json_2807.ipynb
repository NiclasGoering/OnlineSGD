```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

def plot_generalization_error(json_path):
    """
    Loads experiment results from a JSON file and plots the generalization error
    vs. training set size, color-coded by kappa_0.

    Args:
        json_path (str or Path): The full path to the training_results.json file.
    """
    json_path = Path(json_path)
    if not json_path.exists():
        print(f"Error: The file was not found at the specified path:\n{json_path}")
        print("Please make sure the path is correct.")
        return

    # Load the data from the JSON file
    with open(json_path, 'r') as f:
        data = json.load(f)

    # Convert the list of dictionaries to a pandas DataFrame for easy manipulation
    df = pd.DataFrame(data)

    # --- Plotting ---
    plt.style.use('seaborn-v0_8-whitegrid')
    fig, ax = plt.subplots(figsize=(12, 8))

    # Get the unique kappa_0 values to assign a color to each
    unique_kappas = sorted(df['kappa_0'].unique())
    
    # Create a color palette
    palette = sns.color_palette("viridis", n_colors=len(unique_kappas))

    # Plot a separate line for each kappa_0 value
    for i, kappa in enumerate(unique_kappas):
        # Filter the DataFrame for the current kappa_0
        subset = df[df['kappa_0'] == kappa].copy()
        
        # Sort by training set size to ensure the line connects points correctly
        subset.sort_values('P', inplace=True)
        
        # Plot P vs. test_error_01 for the current kappa
        ax.plot(
            subset['P'], 
            subset['test_error_01'], 
            marker='o', 
            linestyle='-', 
            color=palette[i],
            label=f' = {kappa}'
        )

    # --- Formatting the Plot ---
    ax.set_xlabel("Training Set Size (P)", fontsize=14)
    ax.set_ylabel("Generalization Error (Test 0-1 Error)", fontsize=14)
    ax.set_title("Generalization Error vs. Training Set Size for Different ", fontsize=16)
    
    # Use log scales for both axes to better visualize the trends
    ax.set_xscale('log')
    ax.set_yscale('log')
    
    # Add a legend to identify the lines
    ax.legend(title=" values")
    
    # Add grid for better readability
    ax.grid(True, which="both", ls="--", linewidth=0.5)
    
    plt.tight_layout()
    plt.show()

# --- Main Execution ---
# IMPORTANT: Replace this path with the actual path to your JSON file.
# Example for Linux/macOS: "/home/goring/OnlineSGD/results_multigpu/training_results.json"
# Example for Windows: "C:\\Users\\goring\\OnlineSGD\\results_multigpu\\training_results.json"

json_file_path = "/home/goring/OnlineSGD/empirical_tests/results_2707_d60k4_3/training_results.json" 

plot_generalization_error(json_file_path)
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# --- 1. Configuration ---
# Set the path to your results file
# Make sure this path is correct for your environment.
file_path = '/home/goring/OnlineSGD/empirical_tests/results_3007_d40k4_relu3/training_results.json' 
save_plot_path = 'generalization_error_vs_kappa.png'

# --- 2. Load and Process Data ---
try:
    # Load the JSON file into a pandas DataFrame
    df = pd.read_json(file_path)
    print(f"Successfully loaded {len(df)} records from {file_path}")

    # Handle NaN values: treat NaN in 'test_error_01' as a generalization error of 1.0
    # This typically happens when the training diverges at high temperatures (large kappa_0).
    initial_nan_count = df['test_error_01'].isna().sum()
    if initial_nan_count > 0:
        print(f"Found and replaced {initial_nan_count} NaN values in 'test_error_01' with 1.0.")
    df['test_error_01'] = df['test_error_01'].fillna(1.0)
    
    # Sort the DataFrame by 'kappa_0' to ensure the plot is drawn in the correct order
    df_sorted = df.sort_values(by='kappa_0').reset_index(drop=True)
    
    # Extract the relevant columns for plotting
    kappa_0_values = df_sorted['kappa_0']
    gen_error_values = df_sorted['test_error_01']

    # --- 3. Plot the Results ---
    print("Generating plot...")
    sns.set_theme(style="whitegrid", context="talk")
    fig, ax = plt.subplots(figsize=(12, 8))

    # Create the plot with markers and a connecting line
    ax.plot(kappa_0_values, gen_error_values, marker='o', linestyle='-', markersize=8, label='Empirical Results')

    # --- 4. Formatting the Plot ---
    # Set both axes to a logarithmic scale
    ax.set_xscale('log')
    ax.set_yscale('log')

    # Set titles and labels with LaTeX for a professional look
    ax.set_title('Generalization Error vs. Noise Parameter $_0$', fontsize=20, pad=20)
    ax.set_xlabel('Noise Parameter $_0$ (Log Scale)', fontsize=16)
    ax.set_ylabel('Generalization Error (Test 0-1 Error)', fontsize=16)

    # Add a grid for better readability, formatted for log scales
    ax.grid(True, which="both", linestyle='--', linewidth=0.5)
    
    # Optional: Add a horizontal line at 0.5, which is the error for random guessing
    ax.axhline(y=0.5, color='r', linestyle=':', linewidth=2, label='Random Guessing Error (0.5)')
    
    ax.legend(fontsize=14)
    plt.tight_layout()
    
    # Save the figure to a file
    plt.savefig(save_plot_path, dpi=300)
    print(f"Plot saved to '{save_plot_path}'")
    
    # Display the plot in the notebook
    plt.show()

except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found. Please check the path.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
